```python
import multiprocessing as mp
import time

def worker():
    proc = mp.current_process()
    print(proc.name)
    print(proc.pid)
    time.sleep(5)
    print("SubProcess End")    

def run():
    # main process
    proc = mp.current_process()
    print(proc.name)
    print(proc.pid)
    
    # process spawning
    p = mp.Process(name="SubProcess", target=worker)
    p.start()
    p.join()  # 메인 프로세스가 서브 프로세스를 기다리도록 함
    
    print("MainProcess End")
    
run()
```

```python
def get_hpr(ticker):
    df = pybithumb.get_ohlcv(ticker)
    df = df[df.index.year == 2023]
    num_df = len(df)
    
    df['ma5'] = df['close'].rolling(window=5).mean().shift(1)
    df['range'] = (df['high'] - df['low']) * 0.5
    df['target'] = df['open'] + df['range'].shift(1)
    df['bull'] = df['open'] > df['ma5']

    fee = 0.0032
    df['ror'] = np.where((df['high'] > df['target']) & df['bull'],
                          df['close'] / df['target'] - fee,
                          1)

    df['hpr'] = df['ror'].cumprod()
    df['dd'] = (df['hpr'].cummax() - df['hpr']) / df['hpr'].cummax() * 100
    return df['hpr'].iloc[-2], num_df


tickers = pybithumb.get_tickers()

hprs = []

for ticker in tickers:
    try:
        hpr, num_df = get_hpr(ticker)
        hprs.append((ticker, hpr))
    except:
        #print(f"Ticker {ticker} has no hprs! - {num_df}")
        pass

sorted_hprs = sorted(hprs, key=lambda x:x[1])
print(sorted_hprs[-5:])
```
