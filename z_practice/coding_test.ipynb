#### 05 배열 list

```python
# 배열 선언
arr = [0, 0, 0, 0, 0, 0]
print(arr)
arr = [0] * 6
print(arr)

arr = list(range(6))
print(arr)

arr = [0 for _ in range(6)]
print(arr)
```
